I'll start by getting the range and count of each of the `up_vote` values to see if it is a categorical variable.

```python
print(df.agg(
       {
            "up_votes": ["min", "max"]
        }
    ))
df["up_votes"].groupby(df["up_votes"]).count()
```

Looks like there is a large range of up votes. I think that this just means that few people can up vote a post or thousands of people can up vote posts and anywhere in between.

I will do the same for down_votes.

```python
print(df.agg(
       {
            "down_votes": ["min", "max"]
        }
    ))
df["down_votes"].groupby(df["down_votes"]).count()
```

`down_vote` only contains one value (0), so it would be beneficial to remove this column as it does not vary at all.

# Data Cleaning

Some of the titles look like they have other characters besides those contributing to the name of the title of the article. Therefore, I will need to get rid of these characters.

## Remove '\t', '\r', '\n'

```python
for chr in range(len(df.title)):
  df.title[chr] = df.title[chr].translate({ord('\n'): None})
```
